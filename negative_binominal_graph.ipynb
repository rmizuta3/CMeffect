```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import seaborn as sns
import pystan
pd.set_option("display.max_rows", 101)
pd.set_option("display.max_columns", 101)
%matplotlib inline
```

```python
from scipy.stats import nbinom
plt.figure(figsize=(12,8))
x=range(0,20)
#x=np.linspace(0,10,1000)
rs=[1,5]
ps=[0.2,0.5]
for r in rs:
    for p in ps:
        plt.plot(x, nbinom.pmf(x, r,p),label=f"r={r} p={p}")
plt.xlim(0,20)
plt.legend()
plt.show()
```
